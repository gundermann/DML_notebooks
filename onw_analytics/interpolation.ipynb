Get some datapoints from a known function

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return -x**3 + 4*x**2 -5

x = np.linspace(-100,100, 10)

plt.scatter(x, f(x))
plt.plot()
```

Fit the curve using interpolation (but interpolation works only within the interval)

```python
from scipy.interpolate import interp1d

f_inter = interp1d(x, f(x), bounds_error=False, fill_value="extrapolate")
x_inter = np.linspace(-150,150,300)



plt.plot(x_inter, f_inter(x_inter), label='interpolated')
plt.plot(x, f(x), label='original', linestyle='--')
plt.legend()
plt.plot()
```
